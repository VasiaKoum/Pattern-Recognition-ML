#### 2.1 
Έστω $f(x)=x^TAx+b^Tx$, μια συνάρτηση d μεταβλητών, με A συμμετρικό πίνακα (δηλ.$A=A^T$) διάστασης d x d, και $b∈ℝ^d$. Να υπολογίσετε αναλυτικά βήμα προς βήμα με το χέρι, την παράγωγο της $f(x)$ ως προς το διάνυσμα $x∈ℝ^d$ παραθέτωντας όλα τα ενδιάμεσα βήματα.

Η παράγωγος της συνάρτησης $f(x)$ είναι:
$\frac{\partial f(x)}{\partial x} = \frac{\partial (x^TAx+b^Tx)}{\partial x} = $ <br>
$ \frac{\partial (x^TAx)}{\partial x} + \frac{\partial (b^Tx)}{\partial x} = $ <br>
$\frac{\partial (x^TA) +x^TA\partial (x)} {\partial x} + b^T = $ <br>
$\frac{\partial (x^TA)x} {\partial x} +\frac{x^TA\partial (x)} {\partial x} + b^T = $ <br>
$ \frac{\partial (A^Tx)^Tx} {\partial x} +x^TA + b^T = $ <br>
$\frac{(\partial A^Tx)^Tx} {\partial x} +A^Tx + b^T = $ <br>
$(A^T)^Tx+Ax + b^T = $ <br>
$Ax+Ax + b^T = $ <br>
$2Ax+b$

Nα βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution) το ολικό ελάχιστο του παρακάτω προβλήματος βελτιστοποίησης:$$min_X ||A-XB||^2_{F}$$ Να εξηγήσετε με λεπτομέρεια όλη τη διαδικασία σκέψης που θα ακολουθήσετε και να παραθέσετε όλα τα ενδιάμεσα βήματα των πράξεων.

Έχουμε το πρόβλημα βελτιστοποίησης: $min_X ||A-XB||^2_{F}$, οπότε: <br>
$||A-XB||^2_{F} = $ <br>
$(A-XB)^T(A-XB)= $ <br>
$(A^T-B^TX^T)(A-XB)= $ <br>
$A^TA-A^TXB-B^TX^TA+B^TX^TXB= $ <br>
$A^TA-2A^TXB+B^TX^TXB= $ <br>
$A^TA-2A^TXB+X^TXBB^T= $ <br>
Θα παραγωγίσουμε ώστε να βρεθεί το ελάχιστο <br>
$\frac{\partial (A^TA-2A^TXB+X^TXBB^T)}{\partial X} = $ <br>
$(-2A^T+2XBB^T)= $ <br>
Λύνουμε την παραπάνω σχέση ίση με το 0<br>
$(-2A^T+2XBB^T)= 0 =>$ <br>
$2XBB^T=2A^T=>$ <br>
$XBB^T=A^T=>$ <br>
$X=ΑΒ^T(ΒΒ^Τ)^{-1}=>$ <br>
Άρα, $min_X ||A-XB||^2_{F} = ΑΒ^T(ΒΒ^Τ)^{-1}$

## Ερώτημα 3: Gradient Descent
Ο αλγόριθμος gradient descent είναι η πιο διαδεδομένη μέθοδος βελτιστοποίησης στη
μηχανική μάθηση. Σε αυτό το ερώτημα καλείστε να υλοποιήσετε σε Python τον αλγόριθμο
gradient descent για την ελαχιστοποίηση συναρτήσεων δύο μεταβλητών.<br><br>
Θεωρήστε τις παρακάτω συναρτήσεις:
$$f_1([x_1,x_2]^T)=f_1(x_1,x_2)=(x_1-2)^2+(x_2-3)^2$$
$$f_1([x_1,x_2]^T)=f_1(x_1,x_2)=(1-(x_2-3))^2+20((x_1+3)-(x_2-3)^2)^2$$<br>
Nα υλοποιήσετε σε Python και να εκτελέστε τον αλγόριθμο καθόδου κλίσης (gradient descent) για κάθε συνάρτηση. <br>
Αρχικοποιήσετε τις μεταβλητές με 0. Πρώτα, τρέξτε με ένα σταθερό ποσοστό εκμάθησης (learning rate) ίσο με 0.5. Εκτελέστε Τ=10 επαναλήψεις και αναφέρετε την τιμή της
συνάρτησης στο τέλος κάθε βήματος (επανάληψης). Χρησιμοποιώντας τη matplotlib να σχεδιάσετε γραφήματα (ένα για κάθε συνάρτηση) που να απεικονίζουν την τιμή της συνάρτησης (κάθετος άξονας) ως συνάρτηση των επαναλήψεων (οριζόντιος άξονας). <br>
Στη συνέχεια, να επαναλάβετε τη διαδικασία για διαφορετικές τιμές του learning rate και του πλήθους επαναλήψεων (Τ) και αναφέρετε τα αποτελέσματά σας με τη μορφή
γραφημάτων όπως στο παραπάνω σκέλος του ερωτήματος. <br>
* Πόσες επαναλήψεις χρειάζεται ο αλγόριθμος για να συγκλίνει σε στάσιμο σημείο για διαφορετικές τιμές του learning rate
για κάθε συνάρτηση;
* Τι συμβαίνει για μεγάλες τιμές του learning rate;
* Τι συμβαίνει για μικρές τιμές του learning rate;
---------------------------------------------------------------------------------

Θα υπολογίσουμε την παράγωγο των συναρτήσεων, ώστε να εφαρμόσουμε τον εξής τύπο:
$$x_i^j = x_i^{j-1} - \lambda * \frac{\partial f(x_i^{j-1})}{\partial x_i^{j-1}}$$
με $x_i$ το $i-$οστό στοιχείο. 

Η παράγωγος των συναρτήσεων υπολογίστηκε σε πρόχειρο και επιβεβαιώθηκε με τη χρήση της βιβλιοθήκης sympy.

```python
from sympy import symbols, diff

x1, x2 = symbols('x1 x2', real=True)

f = (x1-2)**2+(x2-3)**2
print("Για την πρώτη συνάρτηση: (x1−2)^2+(x2−3)^2")
print("Η παράγωγος ως προς x1 είναι: ", diff(f, x1))
print("Η παράγωγος ως προς x2 είναι: ", diff(f, x2))

f = (1-(x2-3))**2 + 20 * ((x1+3)-(x2-3)**2)**2
print("\nΓια την δέυτερη συνάρτηση: (1-(x2-3))^2+20((x1+3)-(x2-3)^2)^2")
print("Η παράγωγος ως προς x1 είναι: ", diff(f, x1))
print("Η παράγωγος ως προς x2 είναι: ", diff(f, x2))
```

Παρακάτω υπάρχουν δύο συναρτήσεις: **f** και **df**, όπου η f επιστρέφει την τιμή της επιλεγμένης συνάρτησης (whichf=1 ή 2) στις τιμές των x1 και x2 και αντίστοιχα η df επιστρέφει την τιμή της **παραγώγου** της επιλεγμένης συνάρτησης (whichf=1 ή 2) στις τιμές των x1 και x2

```python
def f(x1,x2,whichf):
    if whichf == 1:
        return (x1-2)**2 + (x2-3)**2
    else:
        return (1-(x2-3))**2 + 20 * ((x1+3)-(x2-3)**2)**2    

def df(x1,x2,whichf):
    if whichf == 1:
        return (2*x1-4, 2*x2-6)
    else:
        return (40*x1 - 40*(x2 - 3)**2 + 120, 2*x2 + 20*(12 - 4*x2)*(x1 - (x2 - 3)**2 + 3) - 8)
```

Έπειτα υπάρχουν οι **graddesc** και **convergence**, όπου στην graddesc υπολογίζεται το κάθε βήμα της μεθόδου Gradient Descent και στην convergence εκτελούνται οι $Τ_{max}$ επαναλήψεις με συνθήκη αποκοπής $10^{-3}$ για τον υπολογισμό της σύγκλισης.

```python
def graddesc(x1,x2,l,whichf):
    (f1, f2) = df(x1,x2,whichf)
    return (x1 - l*f1, x2 - l*f2)

def convergence(x1, x2, l, T, whichf):
    xvalues = [(x1, x2, f(x1,x2,whichf))]
    oldf = 0
    for i in range(T):
        (x1,x2) = graddesc(x1,x2,l,whichf)
        if i == 0:
            oldf = f(x1,x2,whichf)
        else:
            if abs(oldf - f(x1,x2,whichf)) < 0.001:
                break
        xvalues.append((x1, x2, f(x1,x2,whichf)))
        oldf = f(x1,x2,whichf)
    return xvalues
```

Τέλος η **graph** εκτυπώνει τα γραφήματα για τις x φορές που θα εκτελεστεί η διαδικασία της gradient descent. Η συνάρτηση ξεκινάει με αρχικές τιμές για T και lambda και ο χρήστης επιλέγει:
*   Πόσες φορές θα εκτελεστεί η διαδικασία της gradient descent,  με διαφορετικές παραμέτρους κάθε φορά
*   Εάν κατά τις επαναλήψεις το Τ θα είναι στατικό ή θα μεταβάλλεται (ξεκινάει με μία αρχική τιμή και σε κάθε επανάληψη αυξάνεται κατά Tincrease)
*   Εάν κατά τις επαναλήψεις το $\lambda$ θα είναι στατικό ή θα μεταβάλλεται (ξεκινάει με μία αρχική τιμή και σε κάθε επανάληψη αυξάνεται κατά lincrease)



```python
def graph(x1, x2, l, T, Tincrease, lincrease, Tstatic, lstatic, x, whichf):
    for i in range(1, int(x)+1):
        plt.figure(figsize=(15, 10))
        values = convergence(x1, x2, l, T, whichf)
        f_val = [x[2] for x in values]
        if x == 1:
            print("f values from every iteration: ", f_val)
        x1_max = max([x[0] for x in values])*10
        x2_max = max([x[0] for x in values])*10
        x1_index = np.linspace(-x1_max, x1_max, 100)
        x2_index = np.linspace(-x2_max, x2_max, 100)
        X1, X2 = np.meshgrid(x1_index, x2_index)

        plt.subplot(2, 2, 1)
        plt.title("For lambda="+str(l)+", T="+str(T))
        cp = plt.contour(X1, X2, f(X1,X2,whichf))
        plt.plot([x[0] for x in values], [x[1] for x in values], 'xr')
        plt.clabel(cp, inline=1, fontsize=10)
        plt.xlabel('X1')
        plt.ylabel('X2')

        plt.subplot(2, 2, 2)
        plt.title("For lambda="+str(l)+", T="+str(T))
        plt.xlabel('Iterations')
        plt.ylabel('F Min')
        plt.plot(f_val)
        plt.plot(f_val, 'ob')

        plt.show()

        if not lstatic:
            l += lincrease
        if not Tstatic:
            T += Tincrease
        if lstatic and Tstatic:
            return
```

Η graph εκτελείται με τις παρακάτω παραμέτρους:
*   **x1** = αρχική τιμή για το x1
*   **x2** = αρχική τιμή για το x1
*   **l** = αρχική τιμή για το $\lambda$
*   **T** = αρχική τιμή για το Τ επαναλήψεις
*   **Tincrease** = ο αριθμός για το επαναληπτικό βήμα για την αύξηση του Τ κατά την εκτέλση των πειραμάτων
*   **lincrease** = ο αριθμός για το επαναληπτικό βήμα για την αύξηση του $\lambda$ κατά την εκτέλση των πειραμάτων
*   **Tstatic** = το Τ θα παραμείνει σταθερό κατά την εκτέλεση των πειραμάτων
*   **lstatic** = το $\lambda$ θα παραμείνει σταθερό κατά την εκτέλεση των πειραμάτων
*   **x** = ο αριθμός των πειραμάτων
*   **whichf** = επιλογή της f ($f_1$ ή $f_2$ από την εκφώνηση)


Αρχικά το ζητούμενο είναι η εκτέλεση των $f_1, f_2$ με αρχικές τιμές 0 , σταθερό learning rate = 0.5 και $T_{max} = 10$ επαναλήψεις.

* Στο πρώτο γράφημα απεικονίζεται η συνάρτηση με χρωματικούς κύκλους όπου ανάλογα το βάθος της 3d συνάρτησης, το χρώμα γίνεται πιο σκούρο. Σε αυτό φαίνονται οι τιμές της f με κόκκινα x και το πραγματικό ελάχιστο να βρίσκεται στο κέντρο του επίκεντρου κύκλου.

* Στο δεύτερο γράφημα απεικονίζεται η πορεία της σύγκλισης όπου στον y-άξονα είναι οι τιμές της συνάρτησης και στον x-άξονα βρίσκονται οι επαναλήψεις.

```python
print("For first function -> (x1−2)^2+(x2−3)^2")
graph(x1=0, x2=0, l=0.5, T=10, Tincrease=10, lincrease=0.1, Tstatic=True, lstatic=True, x=1, whichf=1)
```
